```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import glob
from IPython.display import display

resultadoscsv = glob.glob("*.txt")
resultados = pd.DataFrame()

for result in resultadoscsv:
    m = pd.read_csv(result, sep=';',  header=None,   encoding='utf-8' )
    
    resultados = resultados.append(m, ignore_index=True)

resultados.columns = ["tam_matriz", "abordagem", "tempo_nano", "cpu_usage"]


size = [4,8,16,32,64,128,256,512]
abordagem = ['s','c1','c2','c3']
lista = []

for a in abordagem:
    aux = resultados.loc[resultados['abordagem'] == a]
    for s in size:
        tmp = aux.loc[resultados['tam_matriz'] == s]
        lista.append([a , s, str((tmp.tempo_nano.median() *pow(10,-9)))])
        
saida = pd.DataFrame(lista)
saida.columns = ["abordagem", "tam_matriz", "tempo_nano"]

# saida.to_csv('saida.csv', sep=';')


# 
# c2 = resultados['tempo_nano'].loc[resultados['abordagem'] == 'c2']
# c1 = resultados['tempo_nano'].loc[resultados['abordagem'] == 'c1']
# s = resultados['tempo_nano'].loc[resultados['abordagem'] == 's']
# print(minvalue(c1))

# df=pd.DataFrame({'x': resultados['tam_matriz'], 'y1': np.arange(minvalue(s), maxvalue(s)), 'y2': np.arange(minvalue(c1), maxvalue(c1)), 'y3': np.arange(minvalue(c2), maxvalue(c2)), 'y4': np.arange(minvalue(c3), maxvalue(c3))})

# plt.plot( 'x', 'y1', data=df, marker='o', markerfacecolor='blue')
# plt.plot( 'x', 'y2', data=df, marker='o', markerfacecolor='red')
# plt.plot( 'x', 'y3', data=df, marker='o', markerfacecolor='yellow')
# plt.plot( 'x', 'y4', data=df, marker='o', markerfacecolor='black')


# plt.legend()

fig, ax = plt.subplots()
saida['tempo_nano']=saida['tempo_nano'].astype(float)


for key, grp in saida.groupby(["tam_matriz"]):
    ax = grp.plot(ax=ax, kind='line', x='tam_matriz', y='tempo_nano', c=key, label=key)

plt.legend(loc='best')
plt.show()




# c3 = resultados.loc[resultados['abordagem'] == 'c3']
display(saida)
#saida.plot.line(x='tam_matriz', y='tempo_nano')






# # maprovados =  c3['tempo_nano'].median() * pow(10,-9)

# # print(str(maprovados) + " segundos")
```
