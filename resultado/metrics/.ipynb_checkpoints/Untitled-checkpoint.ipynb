```python
%matplotlib inline

import matplotlib.pyplot as plt
import matplotlib.cm as cm
import pandas as pd
import numpy as np
import glob
from IPython.display import display

resultadoscsv = glob.glob("*.txt")

resultados = pd.DataFrame()

for result in resultadoscsv:
    m = pd.read_csv(result, sep=';',  header=None,   encoding='utf-8' )
    resultados = resultados.append(m, ignore_index=True)

resultados.columns = ["tam_matriz", "abordagem", "tempo_nano"]



lista = []
size = [4,8,16,32,64,128,256,512,1024, 2048]
abordagem = ['s','c1','c2','c3']
for a in abordagem:
    aux = resultados.loc[resultados['abordagem'] == a]
    for s in size:
        if a == 'c1' and s == 2048:
            pass
        tmp = aux.loc[resultados['tam_matriz'] == s]
        lista.append([a , s, str((tmp.tempo_nano.median() *pow(10,-9)))])
        
saida = pd.DataFrame(lista)
saida.columns = ["abordagem", "tam_matriz", "tempo_nano"]

# saida.to_csv('saida.csv', sep=';')

fig, ax = plt.subplots()
saida['tempo_nano']=saida['tempo_nano'].astype(float)


for key, grp in saida.groupby(["abordagem"]):
    ax = grp.plot(ax=ax, kind='line', x='tam_matriz', y='tempo_nano', label=key)

plt.legend(loc='best')
plt.show()



# ================== #


lista = []
size = [4,8,16,32,64,128,256,512,1024, 2048]
abordagem = ['s','c1','c2','c3']
for a in abordagem:
    aux = resultados.loc[resultados['abordagem'] == a]
    for s in size:
        if a == 'c1' and s == 2048:
            pass
        tmp = aux.loc[resultados['tam_matriz'] == s]
        lista.append([a , s, str((tmp.tempo_nano.min() *pow(10,-9))), str((tmp.tempo_nano.max() *pow(10,-9)))])
#display(tempo)
        
tempo = pd.DataFrame(lista)
tempo.columns = ["abordagem", "tam_matriz", "menor_tempo", "maior_tempo"]

fig, ax = plt.subplots()
tempo['menor_tempo']=tempo['menor_tempo'].astype(float)
tempo['maior_tempo']=tempo['maior_tempo'].astype(float)

x = np.arange(10)
ys = [i+x+(i*x)**2 for i in range(10)]
colors = ['Blue', 'Orange', 'Green', 'Red']

for (key, grp), c in zip(tempo.groupby(["abordagem"]), colors):
    grp.plot(ax=ax, kind='scatter', x='tam_matriz', y='menor_tempo', color=c, figsize=(20,10))
    grp.plot(ax=ax, kind='scatter', x='tam_matriz', y='maior_tempo', color=c, secondary_y=True, label=key, figsize=(20,10));

plt.legend(loc='upper left')
plt.show()

```
